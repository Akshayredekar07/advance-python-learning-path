```python
import re 
num=input("Enter mail id: ")
p='[a-zA-Z0-9][a-zA-z0-9._]*@gmail[.]com'
m=re.fullmatch(p, num)
print(num)
if m is not None:
    print('Valid email id')
else:
    print("Invalid email id")
```

```python
# Example 1: Extracting Website Titles
import re
import urllib.request

sites = "google rediff".split()
print(sites)

for s in sites:
    print("Searching...", s)
    u = urllib.request.urlopen("http://" + s + ".com")
    text = u.read()
    title = re.findall("<title>.*</title>", str(text), re.I)
    print(title[0])
```

```python
# Example 2: Extracting Phone Numbers from a Website (redbus.in)
import re
import urllib.request

u = urllib.request.urlopen("https://www.redbus.in/info/contactus")
text = u.read()

numbers = re.findall("[0-9-]{7}[0-9-]+", str(text), re.I)
for n in numbers:
    print(n)
```

```python
# Example 6: Extracting All Links from a Web Page
import re
import urllib.request

u = urllib.request.urlopen("https://www.scaler.com")
text = u.read().decode("utf-8")
links = re.findall("href=['\"](.*?)['\"]", text)
for link in links:
    print(link)
```

```python
import re
import urllib.request

# Function to extract image URLs
def extract_images(url):
    # Fetch the HTML content
    response = urllib.request.urlopen(url)
    html = response.read().decode('utf-8')

    # Regular expression to find all image URLs (both relative and absolute)
    img_pattern = r'<img[^>]+src=["\'](.*?)["\']'
    img_urls = re.findall(img_pattern, html)

    # Convert relative URLs to absolute if needed
    base_url = urllib.parse.urljoin(url, '/')
    img_urls = [urllib.parse.urljoin(base_url, img) for img in img_urls]

    return img_urls

# Example usage
website_url = "https://scaler.com"  # Replace with the target website
images = extract_images(website_url)

# Print extracted image URLs
for img in images:
    print(img)
```

```python
import re
import urllib.request

# Regular expression for extracting academic email IDs
email_pattern = r'[a-zA-Z0-9_.]+@[a-z]+[.]ac[.]in'

# List of university websites
list_of_sites = [
    'https://jntuh.ac.in/contact-us',
    'https://www.osmania.ac.in/aboutus-contactus.php',
    'https://vit.ac.in/contactus',
    'https://mkuniversity.ac.in/new/Quicklinks/Help_line',
    'https://sgbau.ac.in/pages/contact.aspx',
    'https://www.iitm.ac.in/contactus',
    'https://www.iitb.ac.in/en/contact-iit-bombay',
    'https://www.iitkgp.ac.in/contact',
]

def extract_emails(site):
    print(f'\nSearching website: {site}')
    print('#' * 60)

    try:
        # Fetch the HTML content
        resp = urllib.request.urlopen(site)
        html_bytes = resp.read()
        html = html_bytes.decode('utf-8', errors='ignore')  # Decode safely

        # Extract email IDs
        email_ids = set(re.findall(email_pattern, html))

        # Print extracted email IDs
        if email_ids:
            print("\nExtracted Email IDs:")
            for email in email_ids:
                print(email)
        else:
            print("No email IDs found.")

    except Exception as e:
        print(f"Error accessing {site}: {e}")

# Process each website
for site in list_of_sites:
    extract_emails(site)
    print("\n" + "=" * 80 + "\n")
```
